```python
def read_document(file_path):
  with open(file_path,'r',encoding='utf-8') as file:
    return file.readlines();

def map_function(document):
  tokens = document.strip().split()
  return [(token,1) for token in tokens]
```

```python
documents = read_document('/content/text_10k.txt')
print(documents)
```

```python
#apply map function to each document
mapped_results=[]

for document in documents:
    mapped_results.extend(map_function(document))
```

```python
mapped_results
```

```python
# Organise mapped results by key word to reduce phase

intermediate ={}
for key,value in mapped_results:
  if key not in intermediate:
    intermediate[key]=[]
  intermediate[key].append(value)

```

```python
print(intermediate)
```

```python
# reduce function ;aggregates count for each word

def reduce_function(items):
  word,counts = items[0],items[1]
  return(word,sum(counts))
```

```python
# reduce phase
reduce_results= map(reduce_function,intermediate.items())
```

```python
print(list(reduce_results))
```

```python
word_counts= list(reduce_results)
most_common=max(word_counts,key=lambda x:x[1])
least_common=min(word_counts,key=lambda x:x[1])

print(f"most common word {most_common}")
print(f"least common word {least_common}")
```
